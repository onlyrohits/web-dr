```python
import numpy as np
import pandas as pd
import json
```

```python
all_topics_dct = {}
```

```python
questions = pd.read_csv(r'C:\Users\vinit\Desktop\JupyterNotebok\web-dr\webmd_csv\webmd-question.csv')
```

```python
for i in range(len(questions)):
    qtopics = str(questions['Question Topic Id'][i])
    qtopics_list = qtopics.split(',')
    temp = []
    for w in qtopics_list:
        temp.append(w.strip())
    qtopics_list = temp
    for j in range(len(qtopics_list)):
        if all_topics_dct.get(qtopics_list[j],-1)==-1:
            all_topics_dct[qtopics_list[j]] = [questions['Question Id']]
        else:
            all_topics_dct[qtopics_list[j]].append(questions['Question Id'])
```
