```python
from itertools import islice

def take(n, iterable):
    "Return first n items of the iterable as a list"
    return list(islice(iterable, n))

t_q_100 = take(100, topic_q_sorted.items())
```

```python
df_100 = pd.DataFrame(t_q_100)
```

```python
topic_member = dict()
for i in range(df_100.shape[0]):
    topic = df_100.iloc[i,0]
    answer_id = []
    temp_vote_dict = {}
    temp_help_dict = {}
    temp_answer_count = {}
    for qid in df_100.iloc[i,1]:
        member_votes = {}
        member_help = {}
        member_ans_count = {}
        d = np.array(df_answer.loc[df_answer['Question Id'] == qid])
        answer_id.extend(d[:,4])
        answer_count = d[:,4]
        members = d[:,5]
        helpful_num = d[:,3]
        votes = d[:,8]
        for j in range(len(members)):
            member_votes.update({members[j]:votes[j]})
            member_help.update({members[j]:helpful_num[j]})
            member_ans_count.update({members[j]:answer_count[j]})
        for key in member_votes.keys():
            if key in temp_vote_dict:
                temp_vote_dict[key]+=member_votes[key]
            else:
                temp_vote_dict[key]=member_votes[key]
        for key in member_help.keys():
            if key in temp_help_dict:
                temp_help_dict[key]+=member_help[key]
            else:
                temp_help_dict[key]=member_help[key]
        for key in member_ans_count.keys():
            if key in temp_answer_count:
                temp_answer_count[key]+=1
            else:
                temp_answer_count[key]=1
    topic_member.update({topic:{"answers":answer_id,"mem_votes":temp_vote_dict,"mem_help":temp_help_dict,"mem_ans_count":temp_answer_count}})
print(topic_member)
```

```python
with open("/Users/apple/Documents/DV/member_analysis.json",'w') as fp:
    json.dump(topic_member,fp)
```
